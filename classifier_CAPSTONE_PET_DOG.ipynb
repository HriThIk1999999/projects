**STATISTICAL ANALYSIS**

**Hypothesis Test**

1. **Vaccination Status and Health Condition:** Is there a significant difference in the health condition of pets based on their vaccination status?


**Null Hypothesis(H0) :** Vaccinated and non-vaccinated pets have same health condition.

**Alternate Hypothesis(H1) :** vaccinated pets are healthier than non-vaccinated pets.

**Test Selection:** Since we are dealing with two categorical variables (Vaccination Status and Health Condition), we can use the Chi-Square Test of Independence.


```python
from scipy.stats import chi2_contingency

#create contingency table
contingency_table = pd.crosstab(df['Vaccinated'] , df['HealthCondition'])

#Chi Square test
chi2 , p , dof , expected = chi2_contingency(contingency_table)

print('Contigency Table')
print(contingency_table)
print(f"\nChi-Square Statistic: {chi2}")
print(f"\np-value: {p}")
print(f"\nDegree Of Freedom: {dof}")
print('Expected value')
print("\nExpected Frequencies:")
print(expected)

#Conclusion
alpha = 0.5
if p < alpha:
    print("\nReject the null hypothesis. There is a significant relationship between vaccination status and health condition.")
else:
    print("\nFail to reject the null hypothesis. There is no significant relationship between vaccination status and health condition.")

```

2. **Adoption Likelihood and Previous Ownership:** Do pets that had a previous owner have a higher likelihood of being adopted?

**Null Hypothesis:** Pets with and without a previous owner have the same likelihood of being adopted.

**Alternative Hypothesis:** Pets with a previous owner are more likely to be adopted.

**Test Selection:** Since we are dealing with two categorical variables (PreviousOwner and AdoptionLikelihood), we can use the Chi-Square Test of Independence.

```python
#create contingency table
contingency_table2 = pd.crosstab(df['PreviousOwner'] , df['AdoptionLikelihood'])

#test chi2_contingency
chi2 , p , dof , expected = chi2_contingency(contingency_table2)
print('Contigency Table')
print(contingency_table2)
print(f"\nChi-Square Statistic: {chi2}")
print(f"\np-value: {p}")
print(f"\nDegree Of Freedom: {dof}")
print('Expected value')
print("\nExpected Frequencies:")
print(expected)

#Conclusion
aplha = 0.5
if p < alpha:
    print("\nReject the null hypothesis. There is a significant relationship between previous owner and adoption likelihood.")
else:
    print("\nFail to reject the null hypothesis. There is no significant relationship between previous owner and adoption likelihood.")

```

3. **WeightKg and HealthCondition:** Is there a significant difference in the average weight of pets based on their health condition?

**Null Hypothesis(H0) :** Healthy and non-Healthy pets have same average weight.

**Alternate Hypothesis(H1) :** Healthy and non-Healthy pets have different average weight.

**Test Selection:** We are using T-test here

```python
from scipy.stats import ttest_ind

#splitting healthy and non-healthy based on there weights
Healthy = df[df['HealthCondition'] == 0]['WeightKg']
Non_Healthy = df[df['HealthCondition'] == 1]['WeightKg']

#perform T-test
t_test , p_value = ttest_ind(Healthy , Non_Healthy)
print(f"t-statistics:{t_test}")
print(f"p-value: {p_value}")

#Conclusion
alpha = 0.5
if p_value<alpha:
  print('Reject Null Hypothesis .')
else:
  print('Failed to reject null hypothesis')
```

**Conclusion**: Healthy and Non-healthy pets have same average weights .

**Alternate Hypothesis (H1):** Theres no any effect of different types of pets being adopted

```python
from scipy import stats
dog = df[df['PetType']=='Dog']['AdoptionLikelihood']
cat = df[df['PetType']=='Cat']['AdoptionLikelihood']
rabbit = df[df['PetType']=='Rabbit']['AdoptionLikelihood']
bird = df[df['PetType']=='Bird']['AdoptionLikelihood']

f_stat , p_value = stats.f_oneway(dog , cat , rabbit , bird)
print(f'f test value is {f_stat}')
print(f'p_value is {p_value}')

alpha = 0.05
if p_value<alpha:
  print('Reject H0')
else:
  print('Failed to reject H0')
```

**SMOTE TECHNIQUE** : The is similar to OverSampling technique where we will be taking minority classes and perform over sample until the data is balanced . But in RandomOverSampling the sampling is done by duplicating the minority class to acheive data balance which is prone to Overfitting. Whereas SMOTE generates data instead of duplicating to balance the dataset. This is very popular algorithm for balancing the data .

The steps which SMOTES follows are :           

1. Train a knn on minority class observation : find each observations 5 closest neighbors

2. Select examples from the minority class at random.

3. Select a neighbor of each examples at random(for interpolation).

4. Extract a random number from 0 to 1.

5. Calculate the new examples as = Original sample - factor (original sample - neighobor)
